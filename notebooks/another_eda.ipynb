```python
import pandas as pd
import seaborn as sns


df = pd.read_csv(r'..\data\parsed_vcf.csv')
df
```

## Аллели 

```python
sns.histplot(data=df, x='Genotype')
```

## Покрытие участков

```python
sns.histplot(data=df, x='Raw read depth', bins=30)
```

## Участки с достаточным покрытием

```python
sns.histplot(data=df[df['Raw read depth'] >= 10], x='Raw read depth', bins=30)
```

## Наличие таргета по годам

```python
target = pd.read_csv(r'..\data\phenotypes.tsv', sep='\t')
weather = pd.DataFrame(target.notna().sum()[1:])
g = sns.barplot(weather, x=weather.index, y=0)
g.set_xlabel('Год наблюдений')
g.set_ylabel('Количество наблюдений');
```
